### Filtering pandas data

### Pandas joining 2 dataframes

###### There are many ways we can join dataframes based on your requirement like `left`,` right`,` inner`, `outer` . The `how` argument takes care how you want join the table .Make sure column data types are simillar for both the dataframes
###### Simillarly,  you can mention which  columns  needs to be joined based on availability. 
###### if columns are found in both the dataframes, you can use  `on` argument
- `pd.merge(left_dataframe, right_dataframe, on = ['key_1','key_2','key_3'], how = 'left')`

###### if column names are different use `left_on` and `right_on`
- `pd.merge(right_dataframe, right_dataframe, left_on = ['key_1','key_2','key_3'], right_on = ['key_4','key_5','key_6'], how = 'inner')`


###### In case you have same column names in both the dataframes , the resulting dataframe will  have _x and _y  suffixes  for same column names. This default behaviour can be modified by  using `suffixes` argument
- `pd.merge(right_dataframe, right_dataframe, left_on = ['key_1','key_2','key_3'], right_on = ['key_4','key_5','key_6'], how = 'inner',suffixes=('_left','_right'))`

### Dropping duplicates in pandas

- Dropping duplicates . Below syntax will drop duplicates and considers all columns while removing duplicates

- `df.drop_duplicates()`

- Below syntax will drop duplicates as well. But will check duplicates based on the columns mentioned in the subset argument

- `df.drop_duplicates(subset=['col_1','col_2','col_3'])`

- Determines which duplicates (if any) to keep. - first : Drop duplicates except for the first occurrence. - last : Drop duplicates except for the last occurrence. - False : Drop all duplicates.Argument to be used is `keep`

- Whether to drop duplicates in place or to return a copy. Argument to be used is `inplace`

- `df.drop_duplicates(subset=['col_1','col_2','col_3'],keep='first', inplace=True)` 
- `df.drop_duplicates(subset=['col_1','col_2','col_3'],keep='last', inplace=True)`
